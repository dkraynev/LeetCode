Given an m x n integer matrix matrix, if an element is 0, set its entire row and column to 0's.

You must do it in place.

```python
class Solution:
    def setZeroes(self, matrix):
        """
        Do not return anything, modify matrix in-place instead.
        """
        m = len(matrix) # number of rows
        n = len(matrix[0]) # number of columns
        check = max(m,n) # Number of checkings
        diagonal = [3 for i in range(check)]

        # We are going through diagonal (1,1), (2,2) -> (n, n) and
        # Record 4 sitiations
        # 0 - zeroes in rows and colomns:
        # 1 - zeroes only in rows
        # 2- zeroes only in colomns
        # 3 - no zeroes in rows and colomns
        # For this we go in row and colomn (1,1), (2,2) ... (n, n) respectodly
        # and try to find zeroes
        # After that we will go in table again and record zeroes accoarding records
        # We need list size n for records or
        # We can store records also in соответствующих items table in other format
        # for example, in text - "0", "1", "2" or don't change item for situation #3
        for x in range(check): # x - Текущий счетчик начала (первый элемент для текущией строки и для текущего столбца)
            # Проверяем, есть ли в строках 0, если есть, то условие True
            temp_cond_row = x < m or (x >= m and m >= n)
            print(f"Row start: x={x}, n={n}, m={m}, x < m or (x >= m and m >= n: {temp_cond_row}")
            if x < m or (x >= m and m >= n):
                print("Find Conditon for Row:")
                condition_row = False
                for i in range(0, n):
                    print(f"i={i}, x={x}, matrix[x][i]={matrix[x][i]}")
                    if matrix[x][i] == 0:
                        print("Condition_row is True!")
                        print()
                        condition_row = True
                        break

            # Проверяем колонку
            temp_cond_colomn = x < n or (x >= n and n >= m)
            print(f"Colomn start: x={x}, n={n}, m={m}, x < n or (x >= n and n >= m): {temp_cond_colomn}")

            if x < n or (x >= n and n >= m):
                print("Find Conditon for Colomn:")
                condition_colomn = False
                for i in range(0, m):
                    print(f"i={i}, x={x}, matrix[i][x]={matrix[i][x]}")
                    if matrix[i][x] == 0:
                        print("Condition_colomn is True!")
                        print()
                        condition_colomn = True
                        break

            # Record conditons
            if condition_row and condition_colomn:
                diagonal[x] = 0 # Row and colomn have zeroes
            if condition_row and not condition_colomn:
                diagonal[x] = 1 # Only row has zeroes
            if not condition_row and condition_colomn:
                diagonal[x] = 2 # Only colomn has zeroes

            print(f"X={x}, Row={condition_row}, Colomn={condition_colomn}, diagonal[x]={diagonal[x]}")
            print()
            print()

        print("RECORD START:")
        # Record zeroes
        for x in range(check): # x - Текущий счетчик начала (первый элемент для текущией строки и для текущего столбца)
            print(f"X={x}")
            # Record rows
            if diagonal[x] == 0 or diagonal[x] == 1:
                if x < m:
                    print(f"Condition Rows. x={x}, m={m}")
                    for i in range(0, n):
                        print(f"Before record: i={i}, x={x}, matrix[x][i]={matrix[x][i]}")
                        matrix[x][i] = 0
                        print(f"After record: i={i}, x={x}, matrix[i][x]={matrix[x][i]}")
                    print()

            # Record colomns
            if diagonal[x] == 0 or diagonal[x] == 2:
                if x < n:
                    print(f"Condition Colomns. x={x}, n={n}")
                    for i in range(0, m):
                        print(f"Before record: i={i}, x={x}, matrix[i][x]={matrix[i][x]}")
                        matrix[i][x] = 0
                        print(f"After record: i={i}, x={x}, matrix[i][x]={matrix[i][x]}")
                    print()

        # No return
```
